```python
# import data from csv file
import pandas as pd
data = pd.read_csv('bts_army_fanbase_evo.csv')
data.head()
                   
```

```python
# convert data from wide to long format
data_long = pd.melt(data, id_vars = ['Date'], var_name = 'platforms', value_name = 'number_of_fans')
data_long.head()
```

```python
# Convert the 'Week' column to a proper datetime format
data['Date'] = pd.to_datetime(data['Date'], format='%d/%m/%Y')

# Check the updated data
data.head()
```

```python
# export data to new csv file
data_long.to_csv('bts_army_fanbase_evo_long.csv', index = False)
```
