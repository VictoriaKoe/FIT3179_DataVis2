# Search Artists

```python
endpoint_url = "https://api.spotify.com/v1/search?"
```

```python
search = "BTS"
type = 'artist' 
```

```python
query = f'{endpoint_url}'
query += f'&q={search}'
query += f'&type={type}'
```

```python
response = requests.get(query,                                           # get request
                   headers={"Content-Type":"application/json", 
                            "Authorization":f"Bearer {token}"})  
json_response = response.json()                                           # as a json file
```

```python
json_response
```

```python
artist_id = []
artist_url = []
name_list = []                                         
genres = []
followers = []
popularity = []


for i in range(len(json_response['artists']['items'])):
    if json_response['artists']['items'][i] is not None:
        artist_id.append(json_response['artists']['items'][i]['id'])
        artist_url.append(json_response['artists']['items'][i]['external_urls']['spotify'])
        name_list.append(json_response['artists']['items'][i]['name'])
        genres.append(json_response['artists']['items'][i]['genres'])
        followers.append(json_response['artists']['items'][i]['followers']['total'])
        popularity.append(json_response['artists']['items'][i]['popularity'])
```

```python
artist_search_result = pd.DataFrame()
artist_search_result['artist_id'] = artist_id
artist_search_result['artist_url'] = artist_url
artist_search_result['artist_name'] = name_list
artist_search_result['artist_genres'] = genres
artist_search_result['popularity'] = popularity
artist_search_result['total_followers'] = followers
artist_search_result
```

# Get Songs

```python
endpoint_url = "https://api.spotify.com/v1/albums/"

songs_df = pd.DataFrame()

for i in range(len(all_albums_df)):
    album_id = all_albums_df['id'][i]
    
    query = f'{endpoint_url}'
    query += f'{album_id}/tracks'
    
    response = requests.get(query,                                           # get request
                   headers={"Content-Type":"application/json", 
                            "Authorization":f"Bearer {token}"})  
    json_response = response.json()                                           # as a json file
    
    track_id = []
    track_number = []
    duration_ms = []
    track_name = []
    track_preview = []

    for i in range(len(json_response['items'])):
        if json_response['items'][i] is not None:
            track_id.append(json_response['items'][i]['id'])
            track_name.append(json_response['items'][i]['name'])
            track_number.append(json_response['items'][i]['track_number'])
            duration_ms.append(json_response['items'][i]['duration_ms'])        
            track_preview.append(json_response['items'][i]['preview_url'])
    
    temp_df = pd.DataFrame()
    temp_df['track_id'] = track_id
    temp_df['name'] = track_name
    temp_df['album_id'] = album_id
    temp_df['track_number'] = track_number
    temp_df['duration_ms'] = duration_ms
    temp_df['track_preview'] = track_preview
    
    songs_df = songs_df.append(temp_df)
```

```python
songs_df = songs_df.reset_index(drop=True)
songs_df
```

# Get Audio Features

```python
endpoint_url = "https://api.spotify.com/v1/audio-features/"

track_features_df = pd.DataFrame()

for i in range(len(songs_df)):
    track_id = songs_df['track_id'][i]
    
    query = f'{endpoint_url}'
    query += f'{track_id}'
    
    response = requests.get(query,                                           # get request
                   headers={"Content-Type":"application/json", 
                            "Authorization":f"Bearer {token}"})  
    json_response = response.json()                                           # as a json file
    
    danceability = json_response['danceability']
    energy = json_response['energy']
    key = json_response['key']
    loudness = json_response['loudness']
    mode = json_response['mode']
    speechiness = json_response['speechiness']
    acousticness = json_response['acousticness']
    instrumentalness = json_response['instrumentalness']
    liveness = json_response['liveness']
    valence = json_response['valence']
    tempo = json_response['tempo']
    
    temp_df = pd.DataFrame([[track_id, danceability, energy, key, loudness, mode, speechiness, acousticness, instrumentalness, liveness, valence, tempo]], columns=['track_id','danceability', 'energy', 'key', 'loudness', 'mode', 'speechiness', 'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo'], index=[i])
    
    track_features_df = track_features_df.append(temp_df)
```

```python
track_features_df
```

# Merge Datasets

```python
all_df = pd.DataFrame()
all_df = songs_df.join(track_features_df.set_index('track_id'), on='track_id')
all_df
```
